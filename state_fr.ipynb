```python
def main(base_folder,local_path,*args,**kwargs):
    bk.load.current_session_linux(base_folder,local_path)
    neurons, metadata = bk.load.spikes()
    states = bk.load.states(True)
    sleep = bk.load.intervals('sleep_wake')

    states['SLEEP'] = sleep['SLEEP']

    fr = {}
    for state in states:
        fr[state] = []
        for n in neurons:
            fr[state].append(bk.compute.fr(n,states[state]))
    
    fr = pd.DataFrame(fr)
    return pd.concat((metadata,fr),axis=1)
```
