```python
import numpy as np
import bk.load
import matplotlib.pyplot as plt
```

```python
path = '/mnt/electrophy/Ephys/PTSD-Cort/Rat25/Rat25-210416/Rat25_210416_163416/'
```

```python
d = bk.load.digitalin(path+'digitalin.dat')
```

```python
d = d.T
```

```python
ephys = np.memmap(path+'amplifier_analogin_auxiliary_int16.dat',dtype = np.int16).reshape((-1,137))
```

```python
start1 = 40_000
stop1 = 60_000

start2 = -80_000
stop2 = -60_000
fig,ax = plt.subplot_mosaic('''
                            AC
                            BD''')


plt.sca(ax['A'])
ax['A'].plot(d[start1:stop1,0])
plt.ylabel('TTL')
plt.title('Beg of rec')

plt.sca(ax['B'])
ax['B'].plot(ephys[start1:stop1,0])
plt.ylabel('Ephys')
plt.xlabel('Time (samples)')

plt.sca(ax['C'])
plt.title('End of rec')
ax['C'].plot(d[start2:stop2,0])
plt.ylabel('TTL')

plt.sca(ax['D'])
plt.title('End of rec')
ax['D'].plot(ephys[start2:stop2,0])
plt.ylabel('TTL')


plt.suptitle('Example of TTL and Ephys recording sync')
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
a = np.array([[20,5],[18,3.5],[15,2]])
```

```python
reg = ss.linregress(a[:,0],a[:,1])
```

```python
def bac(x,reg):
    return reg.slope * x + reg.intercept
```

```python
print(bac(17.4,reg))
```
