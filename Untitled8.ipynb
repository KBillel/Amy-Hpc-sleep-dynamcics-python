```python
import numpy as np
import bk.load
import neuroseries as nts
import scipy.signal
import matplotlib.pyplot as plt
```

```python
def populationSpikes(neurons):
    pop = np.array([],dtype = np.int64)
    for n in neurons:
        pop = np.hstack((pop,n.times()))
    pop = np.sort(pop)
    pop = nts.Ts(pop)
    return pop
```

```python
def ppWelch(neuron,binsize,nperseg = 5000,noverlap = 2500):
    '''
    Author : BK 
    Data 15/06/21
    
    This function compute the PointProcess PowerSpectra for a spike train using the welch method
    '''
    fs = 1/binsize
    _,b = bk.compute.binSpikes([neuron],binsize,fast = True)
    f,p = scipy.signal.welch(b[0,:],fs,nperseg=nperseg,noverlap=noverlap)
    
    return f,p
```

```python
def plotppWelch(neurons,state):
    pop = populationSpikes(neurons)
    pop = pop.restrict(state)
    if len(pop) == 0: return 0
    
    f,p = ppWelch(pop,0.001)
    plt.loglog(f,p)
    plt.xlim(1,200)
    plt.ylim(pow(10,-7),pow(10,-3))
```

```python
def main(path,arg):
    
    bk.load.current_session_linux(local_path = path)
    neurons,metadata = bk.load.spikes()
    states = bk.load.states()

    stru = ['Hpc','BLA']
    types = ['Pyr','Int']
    plt.figure()
    plt.suptitle(bk.load.session)
    for i,s in enumerate(stru,1):
        plt.subplot(1,len(stru),i)
        for t in types:
            plotppWelch(neurons[(metadata.Region == s) & (metadata.Type == t)],states['sws'])
        plt.title(s)
    plt.legend(types)

    plt.tight_layout()
    plt.show()
```
