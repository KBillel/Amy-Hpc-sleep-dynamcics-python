```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import vBaseFunctions3 as vbf
import assembly as assembly
```

```python
rfolder = '/mnfs/dlt1/data/CocProjFirstData/'

recday = 'mst158-1809-0124' 

b = rfolder + recday + '/' + recday

stages = vbf.LoadStages(b)
units = vbf.LoadUnits(b)
trodes = vbf.LoadUnits(b)

stages
```

```python
s = 4
bs = b+'_'+str(4)
res,clu = vbf.LoadSpikeTimes(bs)
```

```python
actmat,bincentres,cluids = vbf.getActMat(res,clu,binage=int(20/0.8))

tet = 5
thecys = np.load(bs+'.theta.cycles.'+str(tet))
thesmps = vbf.getSamplesWithinEdges(thecys)

binmask = np.in1d(bincentres.astype(int),thesmps)

unitmask = np.array(units['des']=='p1')

actmat_p1_the = actmat[unitmask,:][:,binmask]
cluids_p1 = cluids[unitmask]
```

```python
patterns,_,_ = assembly.runPatterns(actmat_p1_the)

nps = len(patterns)
ncells = len(cluids_p1)

plt.figure(figsize=(25,5))
for pi,pattern in enumerate(patterns):
        plt.subplot(1,nps,1+pi)
        plt.stem(pattern)
        plt.xticks(range(ncells),cluids_p1)
        plt.grid(True)
```

```python
actmat,bincentres,cluids = vbf.getActMat(res,clu,binage=1)

gstd = 20/0.8
actmatS_p1,_ = vbf.MatrixGaussianSmooth(actmat[unitmask],gstd)
actmatS_p1_z = vbf.runthetaz(actmatS_p1,thecys)
```

```python
assemblyact = assembly.computeAssemblyActivity(patterns,actmatS_p1_z)
```

```python
import vPlotFunctions as vp

pi = 2

peak_ = np.argmax(assemblyact[pi])

prange = peak_ + np.array([-300,300])
smps2plot = np.arange(prange[0],prange[1])

vp.plotraster(res,clu,pclu=cluids_p1,prange=prange)
plt.plot(smps2plot,assemblyact[pi][smps2plot]/20)
```

```python
pi = 2

from scipy.signal import find_peaks

thrs = np.std(assemblyact[pi])*5
peaks = find_peaks(assemblyact[pi],height=thrs)[0]
peaks = peaks[np.in1d(peaks,thesmps)]

plt.figure(figsize=(25,5))
plt.plot(assemblyact[pi])
plt.plot(peaks,assemblyact[pi][peaks],'.r')
```

```python
ta,taxis = vbf.triggeredAverage(actmatS_p1_z,peaks)

plt.pcolormesh(taxis,range(ncells),ta,cmap='jet')
plt.colorbar()
plt.grid(True)
```
