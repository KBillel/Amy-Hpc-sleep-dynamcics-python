```python
import numpy as np
import scipy.signal as sp
import matplotlib.pyplot as plt
```

```python
def get_sin(x,amp,fq,phi,noise):
    y = amp*np.sin(x*fq+phi)
    y = y+ noise*np.random.rand(len(y))
    return y
```

```python
x = np.linspace(0,np.pi*20,1000)

y1 = get_sin(x,10,2,0,0)
y2 = get_sin(x,10,2,5,10)
plt.plot(x,y1,x,y2)
```

```python
f,p1 = sp.welch(y1,100)
f,p2 = sp.welch(y2,100)
```

```python
f,c = sp.coherence(y1,y2,100)
```

```python
plt.plot(f,c)
```
