```python
def createsIntervalTimes(base_folder,local_path,*args,**kwargs):
    bk.load.current_session_linux('/mnt/electrophy/Gabrielle/GG-Dataset-Light/',local_path)
    states = bk.load.states()
    
    for state in states:
        states[state]['state'] = state
   
    all_states = pd.concat([states[state] for state in ['sws','Rem','wake','drowsy']])
    translate = [['sws','NREM'],['Rem','REM'],['wake','WAKE'],['drowsy','DROWSY']]
    for t in translate:
        filt = all_states['state'] == t[0]
        all_states.loc[filt,'state'] = t[1]


    sleep = states['Rem'].union(states['sws'])
    sleep['state'] = 'SLEEP'
    

    states['wake_homecage']['state'] = 'WAKE_HOMECAGE'

    sleep_wake = pd.concat((sleep,states['wake_homecage']))

    os.makedirs('Intervals',exist_ok=True)
    os.chdir('Intervals')
    all_states.to_csv('sleep_scoring.csv',index = False)
    sleep_wake.to_csv('sleep_wake.csv',index = False)
    os.chdir(bk.load.path)
```

```python
import bk.load
import scipy.io

import neuroseries as nts

import pandas as pd
```

```python
def create_lap_type(base_folder,local_path):
    bk.load.current_session_linux(base_folder,local_path)
    safe_laps = scipy.io.loadmat(f'{bk.load.session}-LapType.mat')['safelaps'][0][0][0]
    ap_laps = scipy.io.loadmat(f'{bk.load.session}-LapType.mat')['aplaps'][0][0][0]

    safe_laps = nts.IntervalSet(safe_laps[:,0],safe_laps[:,1],time_units='s')
    safe_laps['state'] = 'safe'

    ap_laps = nts.IntervalSet(ap_laps[:,0],ap_laps[:,1],time_units='s')
    ap_laps['state'] = 'aversive'

    df = pd.concat((safe_laps,ap_laps))

    os.makedirs('Intervals',exist_ok=True)
    df.to_csv('Intervals/laps.csv',index = False)
```

```python
bk.load.batch(create_lap_type,local_base='/mnt/electrophy/Gabrielle/GG-Dataset/')
```
